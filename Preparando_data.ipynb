A continuación se eliminan columnas que contienen valores categoricos en especificos que para nuestro analisis no contienen información valiosa y que de hecho en su mayoria son valores nulos (Para revisar el campo que fue eliminado revise la clave en el archivo diccionario_b)

```python
datos_febrero_2022=datos_febrero_2022.drop(columns=['folio', 'control', 'periodo', 'factor','anio','p5_6esp', 'p9_5esp', 'p11_6esp', 'p12_1','p12_2','p12_3','p12_4','p12_5',
                                 'p12_6','p12_7', 'p12_8', 'p12_9', 'p12_9esp', 'p15_5esp', 'p17_6esp', 'p18_1', 'p18_2','p18_3','p18_4','p18_5','p21_5esp',
                                 'p25_6esp','p28_7esp','p32_6esp'] )
```

```python
y=0
for x in datos_febrero_2022.isnull().sum():
  y=y+1
  if x!=0:
    print(f'columna {y} con {x} nulos')
```

```python
datos_febrero_2022=datos_febrero_2022.drop(columns=['p6_6esp'])
```

```python
datos_febrero_2022.shape
```

```python
datos_febrero_2022.describe()
```

Observamos que el valor maximo para la variable nivel (que se refiere al grado de estudios) es 99, mirando el catálogo, 99 significa 'no sabe', pero ya que las demás variables son categoricas ordinales, vamos a sustituir este valor por el valor 0, ya que este representa ningun grado de estudio

```python
datos_febrero_2022.loc[datos_febrero_2022.nivel==99, 'nivel']=0
```

```python
datos_febrero_2022.describe()
```

Revisemos si todos los valores son de tipo int64

```python

for tipo in datos_febrero_2022.dtypes:
  if tipo != 'int64':
    print(tipo)
 
```

```python
from pyod.models.knn import KNN
```

```python
datos_con_posibles_out=datos_febrero_2022.drop(columns=['entidad', 'viv_sel', 'cd', 'sexo', 'nivel', 'edad','num_hog', 'hog_mud', 'n_ren_el', 'cond_act',
                                                        'h_lec',	'mat_lec',	'perslec',	'l_format',	'r_format',	'p_format',	'perslecl'])
```

```python
clf = KNN(0.0009)
clf.fit(datos_con_posibles_out)
y_pred = clf.predict(datos_con_posibles_out)
datos_con_posibles_out[y_pred == 1]
```

Ninguno de los valores que se presentan parecen ser outliers, así que iremos eliminando las columnas que no tengan outliers para así ver el resto de columnas 

```python
datos_con_posibles_out_2=datos_con_posibles_out.drop(columns=datos_con_posibles_out.columns[:10])
```

```python
clf = KNN(0.0009)
clf.fit(datos_con_posibles_out_2)
y_pred = clf.predict(datos_con_posibles_out_2)
datos_con_posibles_out_2[y_pred == 1]
```

Vemos que existe un outlier en la pregunta p7_3, que es el gasto que hizo en el o libros que compro, pero el valor 999999 representa que no sabe, así que vamos a sustituir este valor por la media de dicho gasto

```python
promedio_gasto_libros=datos_febrero_2022[(datos_febrero_2022['p7_3']!=999999)&(datos_febrero_2022['p7_3']!=0)]['p7_3'].mean()
promedio_gasto_libros= int(promedio_gasto_libros)
promedio_gasto_libros
```

```python
datos_con_posibles_out_3=datos_con_posibles_out_2.drop(columns=datos_con_posibles_out_2.columns[:19])
clf = KNN(0.0009)
clf.fit(datos_con_posibles_out_3)
y_pred = clf.predict(datos_con_posibles_out_3)
datos_con_posibles_out_3[y_pred == 1]
```

Obseervamos otro posible outlier en la pregunta p19_3, es similar al anterior punto pero esta vez es gasto en periodicos 

```python
promedio_gasto_periodicos=datos_febrero_2022[(datos_febrero_2022['p19_3']!=999999)&(datos_febrero_2022['p19_3']!=0)]['p19_3'].mean()
promedio_gasto_periodicos=int(promedio_gasto_periodicos)
promedio_gasto_periodicos
```

Se reviso una vez más de posibles outlier pero no se encontrar mas por lo que se procede a sustituir los valores

```python
datos_febrero_2022.loc[datos_febrero_2022.p7_3==999999, 'p7_3']=promedio_gasto_libros
```

```python
datos_febrero_2022.loc[datos_febrero_2022.p19_3==999999, 'p19_3']=promedio_gasto_periodicos
```
